# Método de Newton-Raphson

El método de NR consiste en aproximarse iterativamente a las raíces de una función $f(x)$ a partir de un valor inicial $x_0$
La regla de actualización en cada iteración es 
$x \leftarrow x - dx$, $dx=\frac{f(x)}{f'(x)}$

Y la idea es actualizar el valor de $x$ mientras $dx>\epsilon$ donde $\epsilon$ es una precisión definida por nosotros.


Primero supongamos que sabemos calcular la derivada analíticamente y utilicemos la función
$$f(x) =  (x + 2)(x − 2)(x − 4)$$.
La derivada es
$$\frac{df}{dx}=-4 - 8 x + 3 x^2$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x+2) * (x-2) * (x-4)
def derf(x):
    return -4-8*x+3*x**2
```

```python
xarr = np.linspace(-3, 5, 100)

def NR(x0,epsilon=1e-3):
    dx = f(x0)/derf(x0)
    x=x0
    while abs(dx)>=epsilon:
        dx = f(x)/derf(x)
        x-=dx
        plt.plot(x, f(x), 'o')
    print(x)
```

```python
plt.plot(xarr, f(xarr))
NR(10)
```

```python
plt.plot(xarr, f(xarr))
NR(1)
```

```python
plt.plot(xarr, f(xarr))
NR(-1)
```

Ahora supongamos que no conocemos la derivada analíticamente. Podemos usar el mismo $\epsilon$ para aproximarla numéricamente
$$ \frac{df}{dx} \approx \frac{f(x+\epsilon) - f(x)}{\epsilon}$$. La aproximación numérica presentada acá es la más simple (forward difference). En particular, note que es la misma definición de derivada introducida en cálculo diferencial salvo por la omisión del $\lim_{\epsilon\rightarrow 0}$. La aproximación es útil siempre y cuando $\epsilon$ sea pequeño.


```python
def NR_num(x0,epsilon=1e-10):
    derf = (f(x0+epsilon) - f(x0))/epsilon
    dx = f(x0)/derf
    x=x0
    while abs(dx)>+epsilon:
        derf = (f(x+epsilon) - f(x))/epsilon
        dx = f(x)/derf
        x-=dx
        plt.plot(x, f(x), 'o')
    print(x)
```

```python
plt.plot(xarr, f(xarr))
NR_num(10)
```

```python
plt.plot(xarr, f(xarr))
NR_num(1)
```

```python
plt.plot(xarr, f(xarr))
NR_num(-1)
```
